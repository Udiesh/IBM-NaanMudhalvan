```python
#importing the dataset into the Jupyter Notebook
import pandas as pd
df=pd.read_csv("Assignment-1_Data.csv", delimiter=";")
df
```

```python
column_name = 'Country,,,'  
print(df[column_name])
```

```python
# Removing commas from the 'Country,,,' column and updating the DataFrame
df['Country,,,'] = df['Country,,,'].str.replace(',', '')
df
```

```python
# Renaming the 'Country,,,' column to 'Country' for clarity
df.rename(columns={'Country,,,': 'Country'}, inplace=True)
df
```

```python
# Converting the 'Price' column from comma-separated values to float type
df['Price'] = df['Price'].astype(str).str.replace(',', '.').astype(float)
df
```

```python
# Printing the count of missing values
print("Missing Values:")
print(df.isnull().sum())

#Drop Rows with Missing Values
df.dropna(inplace=True)
```

```python
# Grouping data by 'BillNo' and 'Date', then joining 'Itemname' values
transaction_data = df.groupby(['BillNo', 'Date'])['Itemname'].apply(lambda x: ', '.join(x)).reset_index()

#Drop Unnecessary Columns
columns_to_drop = ['BillNo', 'Date']
transaction_data.drop(columns=columns_to_drop, inplace=True)
transaction_data 
```

```python
# Split the 'Itemname' column into individual items and expand into new columns
items_df = transaction_data['Itemname'].str.split(', ', expand=True)

# Concatenate the original DataFrame with the new items DataFrame
transaction_data = pd.concat([transaction_data, items_df], axis=1)

# Drop the original 'Itemname' column
transaction_data = transaction_data.drop('Itemname', axis=1)

# Display the resulting DataFrame
print(transaction_data.head())
```

```python
# Convert items to boolean columns using one-hot encoding
df_encoded = pd.get_dummies(transaction_data, prefix='', prefix_sep='').groupby(level=0, axis=1).max()

# Save the transaction data with boolean columns to a CSV file
df_encoded.to_csv('transaction_data_encoded.csv', index=False)
```

```python
# Load transaction data with boolean columns into a DataFrame
df_encoded = pd.read_csv('transaction_data_encoded.csv')

from mlxtend.frequent_patterns import apriori, association_rules

# Association Rule Mining with a minimum support of 0.009 and confidence of 0.7
frequent_itemsets = apriori(df_encoded, min_support=0.009, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Display selected association rules with antecedents, consequents, support, and confidence
selected_columns = ['antecedents', 'consequents', 'support', 'confidence']
print("Association Rules:")
print(rules[selected_columns])
```

```python
import plotly.express as px

# Data for visualization
data = {
    'Support': [0.009, 0.009, 0.009, 0.009, 0.009],
    'Threshold': [0.3, 0.4, 0.5, 0.6, 0.7],
    'Number of Rules': [893, 660, 430, 227, 113]
}

# Create a Line Plot using Plotly Express
fig = px.line(data_frame=data, x='Threshold', y='Number of Rules', color='Support',
              labels={'Number of Rules': 'Number of Rules', 'Threshold': 'Threshold'},
              title='Number of Rules vs. Threshold for Support Value of 0.009')

# Customize Line and Marker Styles
fig.update_traces(mode='markers+lines', marker=dict(size=10, line=dict(width=2, color='DarkSlateGrey')),
                  line=dict(width=3))

# Customize Layout for better readability
fig.update_layout(title=dict(font=dict(size=24, family="Arial, sans-serif")),
                  legend=dict(title=dict(font=dict(size=18))),
                  xaxis=dict(title=dict(font=dict(size=20))),
                  yaxis=dict(title=dict(font=dict(size=20))))

# Display the Plot
fig.show()
```

```python
import plotly.express as px

# Convert frozensets to lists for serialization
rules['antecedents'] = rules['antecedents'].apply(list)
rules['consequents'] = rules['consequents'].apply(list)

# Create an interactive scatter plot using Plotly Express
fig = px.scatter(rules, x="support", y="confidence", size="lift",
                 color="lift", hover_name="consequents",
                 title='Market Basket Analysis - Support vs. Confidence',
                 labels={'support': 'Support', 'confidence': 'Confidence'})

# Customize the layout for better visualization
fig.update_layout(
    xaxis_title='Support',
    yaxis_title='Confidence',
    coloraxis_colorbar_title='Lift',
    showlegend=True,
    title_font=dict(size=24, family="Arial, sans-serif"),
    xaxis=dict(title=dict(font=dict(size=20))),
    yaxis=dict(title=dict(font=dict(size=20)))
)

# Show the interactive plot
fig.show()
```

```python
import plotly.express as px

# Combine antecedents and consequents into a single column for each rule
rules['rule'] = rules['antecedents'].astype(str) + ' -> ' + rules['consequents'].astype(str)

# Create a Sunburst chart using Plotly Express
fig = px.sunburst(rules, path=['rule'], values='lift', 
                  title='Market Basket Analysis - Sunburst Chart',
                  color='support', color_continuous_scale='rdylgn')

# Customize the layout for better visualization
fig.update_layout(
    margin=dict(l=0, r=0, b=0, t=40),
    title_font=dict(size=24, family="Arial, sans-serif")
)

# Show the interactive plot
fig.show()
```
